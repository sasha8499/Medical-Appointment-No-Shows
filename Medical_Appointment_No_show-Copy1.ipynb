```python
# Convert to string and remove decimal point from patientId column
df['PatientId'] = df['PatientId'].astype('int64').astype('str')
df.head()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.normalize()  # Extracting only the date (without time)
df.head()
```

```python
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.normalize() # Extracting only the date (without time)
df.head()
```

```python
# Define age bins and labels
bins = [0, 1, 12, 18, 35, 60, 100, 120]
labels = ['Infant', 'Child', 'Teen', 'Young Adult', 'Adult', 'Senior', 'Very Elderly']

# Create a new column for age groups
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df.head()
```

```python
import re

# Remove special characters from neighbourhood names
df['Neighbourhood'] = df['Neighbourhood'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df.head()
```

```python
binary_cols = ['Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'SMS_received']

for col in binary_cols:
    df[col] = df[col].map({0: 'No', 1: 'Yes'}) # Converting 0 and 1 to "YES" and "NO" for better readability
```

```python
df['Handcap'].value_counts()
```

```python
# Converting 0 and 1 to "YES" and "NO" for better readability

df['Handcap'] = df['Handcap'].apply(lambda x: 1 if x > 0 else 0) #Treating anything which is >1 as “handicapped” (combine all into 1)
df.head() 
```

```python
df['Handcap'] = df['Handcap'].map({0: 'No', 1: 'Yes'})
df.head()
```
